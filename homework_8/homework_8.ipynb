## Задание 1

Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:  

1. `car_id = 'А222ВС96'`

Результат: 
`Номер А222ВС валиден. Регион: 96`

2. `car_id = 'АБ22ВВ193'`

Результат: 
`Номер не валиден`

````python
def car_id_check(car_id):
    if re.match(r'[АВЕКМНОРСТУХ]\d{3}[АВЕКМНОРСТУХ]{2}\d{2,3}\b', car_id, flags = re.IGNORECASE):
        print(f'Номер {car_id[0:6]} валиден. Регион: {car_id[6:]}')
    else:
        print('Номер не валиден')
````

````python
car_id = 'А522вВ13'

car_id_check(car_id)
````

````python
def remove_duplicates(some_string):
    new_line = re.sub(r'\b([а-яёА-ЯЁ]+)(?:\s+\1\b)+', r'\1', some_string)
    # начало слова, любое кол-во русских букв, любое кол-во пробельных символов (не включаем в результат), конец слова
    # от 1 до бесконечности повторений этого выражения
    print(new_line)
````

## Задание 3

Напишите функцию, которая будет возвращать акроним по переданной в нее строке со словами.

Примеры работы программы:  

1. `some_words = 'Информационные технологии'`

Результат: 
`ИТ`

2. `some_words = 'Near Field Communication'`

Результат: 
`NFC`

````python
def acronym_composer(some_words):
    acronym = ''.join(re.findall(r'\b[\w|(а-яёА-ЯЁ)]', some_words)).upper()
    # объединить в одну строку без пробелов символ с начала каждого слова и перевести в верхний регистр
    print(acronym)
````

````python
some_words = 'Near Field Communication'

acronym_composer(some_words)
````

## Задание 4

Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

Пример работы программы:  

`emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']`

Результат: 
```
gmail.com: 2
test.in: 1 
ya.ru: 2 
mail.ru: 1
```

````python
def domain_group(emails):
    domain_zone_dict = {} # создать пустой словарь для хранения всех доменных зон
    
    for email in emails:
        if re.search(r'(?<=@)\w+\.\w+', email).group() in domain_zone_dict:
        # после символа @ найти строку "буквы.буквы"; взять найденное значение из RegEx Match Object с помощью метода .group()
            domain_zone_dict[re.search(r'(?<=@)\w+\.\w+', email).group()] += 1
        else:
            domain_zone_dict[re.search(r'(?<=@)\w+\.\w+', email).group()] = 1
        # проверить, была ли доменная зона в словаре и либо добавить 1 к кол-ву, либо записать новую зону
    print(domain_zone_dict)
````

````python
emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']

domain_group(emails)

# При работе с этим заданием возник вопрос по работе с RegEx Match Object
# не получается присвоить переменной значение, полученное при помощи метода .group(), например:
# domain_zone == re.search(r'(?<=@)\w+\.\w+', email).group()
# поэтому пришлось каждый раз вызывать результат re.search с методом .group()
# Можете объяснить почему так и посоветовать понятную статью по RegEx Match Object
````

## Задание 5 (необязательное)

Напишите функцию, которая будет подсчитывать сколько слов начинается на гласные, а сколько на согласные буквы в тексте (текст может быть написан как с использованием букв кириллицы, так и латиницы).

Пример работы программы:  

`some_text = 'Эталонной реализацией Python является интерпретатор CPython, поддерживающий большинство активно используемых платформ. Он распространяется под свободной лицензией Python Software Foundation License, позволяющей использовать его без ограничений в любых приложениях, включая проприетарные.'`

Результат: 
```
Слов на гласные буквы: 9
Слов на согласные буквы: 21 
```

````python
def vowel_consonant_counter(some_text):
    first_symbols = ''.join(re.findall(r'\b[\w|(а-яёА-ЯЁ)]', some_text))
    # объединить в одну строку без пробелов символ с начала каждого слова
    vowels_list = re.findall(r'[aeiouаеёиоуыэюя]', first_symbols, re.IGNORECASE)
    consonant_list = re.findall(r'[bcdfghjklmnpqrstvwxyzбвгджзйклмнпрстфхцчшщъь]', first_symbols, re.IGNORECASE)
    # создать списки гласных и согласных букв путем поиска соответствующих кириллических и латинских символов
    print(f'Слов на гласные буквы: {len(vowels_list)}\nСлов на согласные буквы: {len(consonant_list)}\nЦифр и специальных символов: {len(first_symbols) - len(vowels_list) - len(consonant_list)}')
````

````python
some_text = 'Эталонной реализацией Python является интерпретатор CPython, поддерживающий большинство активно используемых платформ. Он распространяется под свободной лицензией Python Software Foundation License, позволяющей использовать его без ограничений в любых приложениях, включая проприетарные.'

vowel_consonant_counter(some_text)
````
